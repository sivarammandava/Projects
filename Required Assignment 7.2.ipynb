```python
import numpy as np
import pandas as pd
```

```python
babyNames = pd.read_csv("babynames.csv")
```

```python
babyNames.head()
```

```python
filHilary= babynames[(babynames['name'] == 'Hilary')]
filHilary.head(10)
```

```python
countHilary= filHilary.groupby(['year', 'sex'])["n"].sum()
countHilary
```

```python
numNames = (babyNames.groupby(['year','sex']).size())
numNames
```

```python
babyNames['rank'] = babyNames.groupby(['year','sex'])['prop'].rank(ascending = False)
babyNames.head()
```

```python
babyNames[(babyNames['year']==1999) & (babyNames['rank']==1.0)]
```

```python
babyNames.groupby(['name', 'sex'])[['n']].agg(np.median).sort_values(by='n', ascending=False).head(20)
```

```python
babyNamesTop100 = babyNames[babyNames['rank'] <= 100]
babyNamesTop100
```

```python
babyNamesTop100.groupby(['year','sex']).prop.sum()
```

```python
babyNames.loc[babyNames.groupby(['name','sex'])['rank'].idxmin()].sort_values(by = 'rank')
```

```python
top10Names = babyNames[babyNames['rank'] < 10]
top10Names = top10Names.groupby(['name', 'sex']).size().reset_index()
top10Names.columns = ['name', 'sex', 'top10Names']
top10Names.sort_values(by = 'top10Names', ascending = False).head(10)
```
